```python
codon_table = { 
"UUU" : "F", "CUU" : "L", "AUU" : "I", "GUU" : "V",
"UUC" : "F", "CUC" : "L", "AUC" : "I", "GUC" : "V",
"UUA" : "L", "CUA" : "L", "AUA" : "I", "GUA" : "V",
"UUG" : "L", "CUG" : "L", "AUG" : "M", "GUG" : "V",
"UCU" : "S", "CCU" : "P", "ACU" : "T", "GCU" : "A",
"UCC" : "S", "CCC" : "P", "ACC" : "T", "GCC" : "A",
"UCA" : "S", "CCA" : "P", "ACA" : "T", "GCA" : "A",
"UCG" : "S", "CCG" : "P", "ACG" : "T", "GCG" : "A",
"UAU" : "Y", "CAU" : "H", "AAU" : "N", "GAU" : "D",
"UAC" : "Y", "CAC" : "H", "AAC" : "N", "GAC" : "D",
"UAA" : "Stop", "CAA" : "Q", "AAA" : "K", "GAA" : "E",
"UAG" : "Stop", "CAG" : "Q", "AAG" : "K", "GAG" : "E",
"UGU" : "C", "CGU" : "R", "AGU" : "S", "GGU" : "G",
"UGC" : "C", "CGC" : "R", "AGC" : "S", "GGC" : "G",
"UGA" : "Stop", "CGA" : "R", "AGA" : "R", "GGA" : "G",
"UGG" : "W", "CGG" : "R", "AGG" : "R", "GGG" : "G"
}
```

```python
sample = str(input("Please enter the RNA String: "))

chunks = []  # create an empty list to store extracted data

for i in range(0, len(sample), 3):
    piece = sample[i:i+3]  # extract the 3-letter substring
    chunks.append(piece)  # add the extracted piece to the chunks list

protein = []
for chunk in chunks:
    if codon_table[chunk] != "Stop":
        protein.append(codon_table[chunk])
    else:
        break

concatenated_string = "".join(protein)

print(concatenated_string) 

```
