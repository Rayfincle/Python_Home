```python
data = input("Enter your data here, please: ")
```

```python
#data = data.replace(" ", "")
print(data)
```

```python
split_data = data.split('>')
#print(split_data)
del split_data[0]
stripped_data = [entry.rstrip() for entry in split_data] # Super fuckin useful
#print(stripped_data)
split_data = stripped_data
```

```python
new_split_data = []
for i in split_data:
    split_words = i.split(" ",1)
    new_split_data.append(split_words)

print(new_split_data)

data_dict = {}

for sublist in new_split_data:
    key = sublist[0]
    value = sublist[1]
    data_dict[key] = value

print(data_dict)
```

```python
for key, value in data_dict.items():
    prefix = value[:3]  # Extract the first three characters as prefix
    suffix = value[-3:]  # Extract the last three characters as suffix

    for key1, value1 in data_dict.items():
        if key == key1:
            continue

        prefix1 = value1[:3]  # Extract the first three characters as prefix
        suffix1 = value1[-3:]  # Extract the last three characters as suffix

        if suffix == prefix1:
            print(key, key1)
```
