```python
def mortal_rabbits(n,m):
    F = [1,1] + [0] *(n-2)
    
    for i in range(2,n):
        if i < m:
            F[i] = F[i-1] + F[i-2]
        elif i == m:
            F[i] = F[i-1] + F[i-2] -1
        else:
            F[i] = F[i-1] + F[i-2] - F[i-m-1]
    
    print(F)
    

mortal_rabbits(85, 20)
```
