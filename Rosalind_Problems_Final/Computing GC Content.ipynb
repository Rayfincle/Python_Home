```python
data = input("paste your data ")   
```

```python
chunks = []  # create an empty list to store extracted data

start = 0  # initialize start variable to 0
while True:  # create a loop that continues until broken
    start = data.find(">", start)  # find the next ">" symbol in data starting from the current value of start
    if start == -1:  # if ">" symbol is not found, break the loop
        break
    end = data.find(">", start+1)  # find the next ">" symbol after the start symbol in data
    if end == -1:  # if ">" symbol is not found, set end to the length of text
        end = len(data)
    piece = data[start+1:end].strip()  # extract the text between the two ">" symbols and remove any leading/trailing whitespaces
    chunks.append(piece)  # add the extracted piece to the chunks list
    start = end  # set the start value to the end value for the next iteration of the loop

print(chunks)  # print the final list of extracted data chunks
```

```python
from collections import Counter

# Loop over each element in the `chunks` list
for i, x in enumerate(chunks):
    # Count the number of occurrences of each character in `x`
    Answer = Counter(x)
   
    # Define an empty list to store the names
    list_item_names = []
    # Extract the counts as a list and store in `result`
    result = list(Answer.values())
    
    # Extract the counts for each nucleotide and store in separate variables
    A = Answer['A']
    C = Answer['C']
    G = Answer['G']
    T = Answer['T'] 
    
    # Calculate the percentage of GC content in `x`
    Percent_GC = ((C + G) / (A + C + G + T) * 100)

    # Iterate over each item in the list
    for item in chunks:
        # Split the item by the first space character
        name, sequence = item.split(' ', 1)
        # Append the name to the list_item_names list
        list_item_names.append(name)
        
    
    print(list_item_names[i])
    print('{:.6f}'.format(Percent_GC))
    print()
```
